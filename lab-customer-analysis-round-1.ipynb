## 1. Case Study

```python
import pandas as pd
import numpy as np
```

## 2. Get data

```python
#To import the file in Jupyter notebook
file1 = pd.read_csv('files_for_lab/csv_files/file1.csv')
file2 = pd.read_csv('files_for_lab/csv_files/file2.csv')
file3 = pd.read_csv('files_for_lab/csv_files/file3.csv')
```

## 3. Cleaning/Wrangling/EDA

```python
#Now I´m cheking the order of the columns, we can see that they don´t have the same order
print(file1.columns)

print(file2.columns)

print(file3.columns)
```

```python
#I´ll take file 1 as a model to short all the dataframes
file2 = file2[['Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]

file3 = file3.rename(columns = {'State':'ST','Gender':'GENDER'})
file3 = file3[['Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
```

```python
#Now all the columns have the same order.
print(file1.columns)

print(file2.columns)

print(file3.columns)
```

```python
#Next step is to join all the files in one
file = pd.concat([file1, file2, file3], axis=0)
file.shape
```

```python
#This is to became all the clumns in lower case usir a for loop.
cols =[]
for i in file1.columns:
    cols.append(i.lower())
file.columns = cols

file.head()
```

```python
#Cleaning the data droping duplicates and reseting the index
file = file.drop_duplicates()
file = file.reset_index(drop=True)

file.shape
```

```python
file.dtypes
```

```python
#file._get_numeric_data()
file.select_dtypes('object')
#file.select_dtypes('float')
```

```python
#Delete the column education and the number of open complaints from the dataframe.
file = file.drop(['education','number of open complaints'], axis=1)
file.head()
```

```python
#Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type
file['customer lifetime value'] = pd.to_numeric(file['customer lifetime value'],errors= 'coerce')
file.head()

file['customer lifetime value'] = file['customer lifetime value']/100
```

```python
# Customers who have an income of 0 or less.
file = file[file['income']>0]
file.head()
```
